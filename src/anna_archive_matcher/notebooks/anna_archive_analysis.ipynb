# Anna's Archive Book Matching Analysis

This notebook provides interactive analysis of the book matching process between your romance novels dataset and Anna's Archive collections.

## Overview

The matching system uses DuckDB to query Anna's Archive datasets:
- **Elasticsearch Dataset**: Main book collection with metadata
- **AAC Dataset**: Additional book records and metadata  
- **MariaDB Dataset**: Archive metadata and file information

## Features

- High-performance SQL queries with DuckDB
- Multiple matching strategies (exact, fuzzy, author-based)
- MD5 hash extraction for automated downloads
- Comprehensive matching statistics and analysis
